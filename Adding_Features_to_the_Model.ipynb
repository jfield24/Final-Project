```python
from utils.features import *
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
for i in range(2011, 2022):
    if i <=2012:
        link = 'data/' + str(i) + '.xls'
    else:
        link = 'data/' + str(i) + '.xlsx'
    df_temp = pd.read_excel(link)
    df = df.append(df_temp, sort=False)
```

```python
df = df.reset_index()
```

```python
df = df[df.Date.notnull()]
```

```python
df
```

```python
df['Date'] = df.apply(lambda row: datetime.strptime(str(row['Date']), "%Y-%m-%d %H:%M:%S").strftime("%Y/%m/%d"), axis=1)
```

```python
df.reset_index(inplace=True)
```

```python
del (df['index'])
```

```python
df = df[df.Comment == 'Completed']
```

```python
df.loc[:,'best_of_5'] = (df['Best of'] ==5).astype(int)
df['W3'] = pd.to_numeric(df['W3'], errors='coerce')
df['L3'] = pd.to_numeric(df['L3'], errors='coerce')
```

```python
df = df.fillna(0)
```

```python
print(df.dtypes)
```

```python
df_combined = df[['Tournament',  'Date', 'Surface', 'Round']].copy()
```

```python
df_combined.loc[:,'player_1'] = df.apply(lambda row: find_player_1_name(row['Winner'], row['WRank'], row['Loser'], row['LRank']), axis=1)
df_combined.loc[:,'player_1_rank'] = df.apply(lambda row: find_player_1_rank(row['WRank'], row['LRank']), axis=1)
df_combined.loc[:,'player_1_odds'] = df.apply(lambda row: find_player_1_odds(row['AvgW'], row['WRank'], row['AvgL'], row['LRank']), axis=1)
```

```python
df_combined.loc[:,'player_2'] = df.apply(lambda row: find_player_2_name(row['Winner'], row['WRank'], row['Loser'], row['LRank']), axis=1)
df_combined.loc[:,'player_2_rank'] = df.apply(lambda row: find_player_2_rank(row['WRank'], row['LRank']), axis=1)
df_combined.loc[:,'player_2_odds'] = df.apply(lambda row: find_player_2_odds(row['AvgW'], row['WRank'], row['AvgL'], row['LRank']), axis=1)
```

```python
df_combined.loc[:,'result'] = df.apply(lambda row: result(row['WRank'], row['LRank']), axis=1)
```

```python
df_combined = df_combined[df_combined.Tournament == 'Australian Open']
```

```python
df_combined = df_combined[df_combined.Date > '2010/01/01']
```

```python
df_combined = add_features(df_combined, df)
```
